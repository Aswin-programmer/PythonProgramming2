```python
#############################MERGE, JOIN & CONCATENATE DATAFRAMES #######################################
```

```python
df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Keerti', 'Amit', 'Educosys', 'YouTube']
})

df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'course' : ['GenAI', 'CPP', 'HLD', 'LLD']
})
```

```python
inner_result = pd.merge(df1, df2, how = 'inner', on = 'id')#Here the inner actually acts as the index.
inner_result
```

```python
outer_result = pd.merge(df1, df2, how = 'outer', on = 'id')#Here the outer actually as the union
outer_result
```

```python
left_result = pd.merge(df1, df2, how = 'left', on = 'id')#Here the left actually null values from the union. Note : df1=>LEFT, df2=>RIGHT
left_result
```

```python
right_result = pd.merge(df1, df2, how = 'right', on = 'id')#Here the right acutually null values from the union. Note : df1=>LEFT, df2=>RIGHT
right_result
```

```python
#Sometimes the common column could have the different id name(EX: id1 and id2).
df3 = pd.DataFrame({
    'id1' : [1, 2, 3, 4],
    'name' : ['Keerti', 'Amit', 'Educosys', 'YouTube']
})

df4 = pd.DataFrame({
    'id2' : [3, 4, 5, 6],
    'course' : ['GenAI', 'CPP', 'HLD', 'LLD']
})

inner_result = pd.merge(df3, df4, how = 'inner', left_on = 'id1', right_on = 'id2')
inner_result
```

```python
df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Keerti', 'Amit', 'Educosys', 'YouTube']
})

df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'course' : ['GenAI', 'CPP', 'HLD', 'LLD']
})

result1 = pd.concat([df1, df2], axis = 0, ignore_index = True)#Axis(0) indicates that vertical concat. Ignore the default index and create your own index.
result1
```

```python
result2 = pd.concat([df1, df2], axis = 1, ignore_index = True)#Axis(1) indicates that horizon concat. Ignore the default index and create your own index.
result2
```

```python
result3 = pd.concat([df1, df2], axis = 0, keys = ['df1', 'df2'])#This helps to index the data set.
result3
```
