```python
x = [1, 2, 3, 4]
y = [67, 60, 54, 80]

plt.plot(x, y)
plt.title("Marks correspponding to number of hours of study")
plt.xlabel("No of hours of study")
plt.ylabel("Marks")

plt.xticks([1, 2, 3, 4])
plt.yticks([1, 50, 100])

plt.show()
```

```python
x = [1, 2, 3, 4]
dsa_marks = [67, 60, 54, 80]
ml_marks = [60, 50, 34, 10]

#You can also add ,color = 'red' or color = '#000000', Similarly we can also give linewidth = 2 and marker = '*'
plt.plot(x, dsa_marks, label = 'dsa_marks', color = 'black', linewidth = 3, marker = '*', linestyle = '--', markersize = 20)
plt.plot(x, ml_marks, label = 'ml_marks')

plt.title("Marks correspponding to number of hours of study", fontdict = { 
    'fontname' : 'times new roman',
    'fontweight' : 10,
    'color' : 'green'
})
#All the things that are being done in the tile can also applied to the xlabel and ylabel.
plt.xlabel("No of hours of study")
plt.ylabel("Marks")

plt.xticks([1, 2, 3, 4])

plt.legend(loc = 'upper right')

plt.savefig('marks.png')

plt.savefig('marksWithLowDPI.png', dpi = 1000, transparent = 1)#This line has to be called before the plt.show() orelse the empty figure will be exported.

plt.show()
```

```python
##GOOD EXAMPLE
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2) #(StartingPoint, EndingPoint, IncrementAmount)

print(t)

#red dashes, blue squares, green triangles
plt.plot(t, t, color = 'red', linestyle = '--')

plt.plot(t, t**2, color = 'blue', linestyle = '', marker = 's')

plt.plot(t, t**3, color = 'green', linestyle = '', marker = '^')

plt.show()
```

```python
t = np.arange(0., 5., 0.2) #(StartingPoint, EndingPoint, IncrementAmount)

plt.figure()

plt.subplot(1, 3, 1)
plt.plot(t, t, color = 'red', linestyle = '--')
plt.title('Linear Subplot')#Similarly we can do it for xlabel, ylabel and etc.

plt.subplot(1, 3, 2)
plt.plot(t, t**2, color = 'blue', linestyle = '', marker = 's')
plt.title('Quadratic Subplot')

plt.subplot(1, 3, 3)
plt.plot(t, t**3, color = 'green', linestyle = '', marker = '^')
plt.title('Cubic Subplot')

plt.suptitle('Different functiions of t')
plt.tight_layout()

plt.show()
```

```python
# asex - x axis, y axis, lables, tickers

x = np.linspace(0, 10, 100)#Between 10 to 100, Generate 100 evenly distributed points.
print(x)

#Before we saw subplot not now we actually see the subplots.
fig, axis = plt.subplots(2, 3)#(no of rows, no of columns)

axis[0,0].plot(x, np.sin(x))
axis[0,0].set_title('Sin Func')
axis[0,0].set_xlabel('x')
axis[0,0].set_ylabel('y')

axis[0,1].plot(x, np.cos(x))
axis[0,1].set_title('Cos Func')
axis[0,1].set_xlabel('x')
axis[0,1].set_ylabel('y')

axis[0,2].plot(x, np.tan(x))
axis[0,2].set_title('Tan Func')
axis[0,2].set_xlabel('x')
axis[0,2].set_ylabel('y')

axis[1,0].plot(x, np.exp(x))
axis[1,0].set_title('Exp Func')
axis[1,0].set_xlabel('x')
axis[1,0].set_ylabel('y')

axis[1,1].plot(x, np.sqrt(x))
axis[1,1].set_title('Sqrt Func')
axis[1,1].set_xlabel('x')
axis[1,1].set_ylabel('y')
1
axis[1,2].plot(x, np.log(x))
axis[1,2].set_title('Log Func')
axis[1,2].set_xlabel('x')
axis[1,2].set_ylabel('y')
1

plt.tight_layout()
```

```python
#Bar Graph
courses = ['GenAi', 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]

plt.bar(courses, students)

plt.show()
```

```python
#Bar Graph
courses = ['GenAi', 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]

plt.bar(courses, students, width=0.5, align = 'center', edgecolor = 'black', linewidth = 3, color = 'red', hatch = '*')

plt.show()
```

```python
#Bar Graph
courses = ['GenAi', 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]
hatches = ['/', '*', 'o', '\\', '|', '-']

bars = plt.bar(courses, students, width=0.5, align = 'center', edgecolor = 'black', linewidth = 3, color = 'red', hatch = '*')

for bar, hatch in zip(bars, hatches):
    bar.set_hatch(hatch)

plt.xlabel('Courses')
plt.ylabel('No of students')
plt.title('No of students enrolled in courses')

plt.show()
```

```python
#Bar Graph
courses = ['GenAi', 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]
hatches = ['/', '*', 'o', '\\', '|', '-']
bar_colors = ['red', 'blue', 'magenta', 'yellow', 'orange', 'green']

bars = plt.bar(courses, students, width=0.5, align = 'center', linewidth = 3)

for bar, hatch, bar_color in zip(bars, hatches, bar_colors):
    bar.set_color(bar_color)
    bar.set_hatch(hatch)
    bar.set_edgecolor('black')

plt.xlabel('Courses')
plt.ylabel('No of students')
plt.title('No of students enrolled in courses')

plt.show()
```

### Histogram Plot ###

```python
ages = [34, 23, 78, 54, 67, 45, 21, 98, 56, 78]

ages.sort()

print(ages)

plt.hist(ages, bins = 5)

plt.show()
```

```python
ages = [34, 23, 78, 54, 67, 45, 21, 98, 56, 78]

ages.sort()

print(ages)

plt.hist(ages, bins = 5, orientation = 'horizontal', edgecolor = 'black', color = 'green')

plt.xlabel('Count')
plt.ylabel('Ages')
plt.title('Ages Plot')

plt.show()
```

```python
#Pie plot

courses = ['GenAi', 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]

plt.pie(students, labels = courses)

plt.show()
```

```python
#Pie plot

courses = ['GenAi', 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]
explode = [0.2, 0, 0, 0, 0, 0]
pie_colors = ['red', 'blue', 'magenta', 'yellow', 'orange', 'green']

plt.pie(students, labels = courses, shadow = True, explode = explode, colors = pie_colors, autopct = '%.2f%%')

plt.legend(loc = 'upper right')

plt.show()
```

```python
#3D Scatter Plots

ax = plt.axes(projection = '3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y ,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_title('3D Scattered Plot')

plt.show()
```

```python
#3D Curve plot

ax = plt.axes(projection = '3d')

x = np.arange(0, 100, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)

plt.show()
```
