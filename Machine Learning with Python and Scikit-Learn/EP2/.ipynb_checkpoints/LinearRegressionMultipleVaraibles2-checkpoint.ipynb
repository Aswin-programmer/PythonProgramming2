```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
print(df)
```

```python
df.loc[2, 'bedrooms'] = np.nan
print(df)
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
print(df)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area' ,'bedrooms', 'age']], df['price'])
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
print(reg.predict([[3000, 3, 40]]))
```

```python
print(reg.predict([[2500, 4, 5]]))
```

```python
df2 = pd.read_csv('hiring.csv')
print(df2)
```

```python
df2temp = df2.copy()
print(df2temp)
```

```python
#preprocesing
df2temp['experience'] = df2temp['experience'].fillna('zero')
print(df2temp)
```

```python
df2temp.rename(columns = {
    'test_score(out of 10)' : 'test_score',
    'interview_score(out of 10)' : 'interview_score'
}, inplace = True)
print(df2temp)
```

```python
df2temp['test_score'] = df2temp['test_score'].fillna(df2temp['test_score'].median())
print(df2temp)
```

```python
from word2number import w2n
# testing
print(w2n.word_to_num('two point three'))
```

```python
def changeWordToNumber(x):
    return w2n.word_to_num(x)

df2temp['experience'] = df2temp['experience'].apply(changeWordToNumber)

print(df2temp)
```

```python
# Training the machine learning model
reg = linear_model.LinearRegression()
reg.fit(df2temp[['experience', 'test_score', 'interview_score']], df2temp['salary($)'])
```

```python
print(reg.predict([[2, 9, 6]]))
```

```python
print(reg.predict([[12, 10, 10]]))
```
