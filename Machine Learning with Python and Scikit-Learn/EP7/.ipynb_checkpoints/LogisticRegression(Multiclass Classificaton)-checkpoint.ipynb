```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
print(digits['data'][0])
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits['images'][i])
```

```python
print(digits['target'][0 : 5])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits['data'], digits['target'], test_size = 0.2, random_state = 10)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test,  y_test)
```

```python
print(model.predict([digits['data'][5]]))
```

```python
plt.gray()
plt.matshow(digits['images'][5])
```

```python
plt.gray()
plt.matshow(digits['images'][6])
print(model.predict([digits['data'][6]]))
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10, 7))
sn.heatmap(cm, annot = True)# When annot=True, each cell in the heatmap will display the corresponding value from cm (the matrix you're plotting).
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
