```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
print(df)
```

```python
#Draw a scatter plot to analyse the data
plt.scatter(df.loc[:, 'area'], df.loc[ : ,'price'], color = 'red', marker = '+')
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
```

```python
reg = linear_model.LinearRegression()# y = mx + c
reg.fit(df[['area']], df['price'])

# df['area']              → Shape: (5,)   → 1D (Series)
# [df['area']]            → Shape: (1, 5) → 2D but wrong orientation (1 row, 5 columns)
# df[['area']]            → Shape: (5, 1) → ✅ 2D, 5 rows, 1 column (correct for sklearn)
```

```python
print(reg.coef_) # slope or gradient {m}.

print(reg.intercept_) # intercept {c}.
```

```python
print(reg.predict([[3300]]))
```

```python
# y = mx + c
print(135.78767123 * 3300 + 180616.43835616432)
```

```python
#Draw a scatter plot to analyse the data
plt.scatter(df.loc[:, 'area'], df.loc[ : ,'price'], color = 'red', marker = '+')
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.plot(df.loc[:, 'area'], reg.predict(df[['area']]), color = 'green')
plt.show()
```

```python
df2 = pd.read_csv('canada_per_capita_income.csv')
print(df2.head())
```

```python
df2temp = df2.copy()
```

```python
df2temp.rename(columns = {'per capita income (US$)' : 'per_capita_income'}, inplace = True)
df2temp.head()
```

```python
plt.scatter(df2temp['year'], df2temp['per_capita_income'], color = 'red', marker = '+')
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df2temp[['year']], df2temp['per_capita_income'])
```

```python
reg.predict([[2020]])
```

```python
plt.scatter(df2temp['year'], df2temp['per_capita_income'], color = 'red', marker = '+')
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.plot(df2temp['year'], reg.predict(df2temp[['year']]), color = 'green')
plt.show()
```
