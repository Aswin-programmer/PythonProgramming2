````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
````

````python
df = pd.read_csv('insurance_data.csv')
print(df)
````

````python
plt.scatter(df['age'], df['bought_insurance'], color = 'red', marker = '+')
plt.show()
````

````python
df.shape
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df['bought_insurance'], test_size = 0.1, random_state = 10)
````

````python
print(X_train)
print(X_test)
````

````python
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
````

````python
print(model.predict(X_test))
````

````python
model.score(X_test, y_test)
````

````python
model.predict_proba(X_test)
````

````python
import seaborn as sns
# ðŸ”¢ What is a Count Plot?
# âœ… Definition:
# A count plot is a bar plot that shows the number of occurrences of each category in a categorical feature (like salary, Department, etc.).

# ðŸŽ¯ Purpose:
# To visualize the distribution of a categorical variable.

# To compare categories directly.

# With hue, to compare across two categories (e.g., how many people left vs didn't for each salary level).

# âœ… Example Use Case
# Problem:
# "I want to know how many employees with each salary level have left the company."

# ðŸ”§ Syntax of Seaborn countplot
# sns.countplot(x='column_name', data=df)
# Optional:
# sns.countplot(x='column', hue='another_column', data=df)
````

````python
plt.figure()
sns.countplot(x = 'salary', hue = 'left', data = df2)
plt.title('Employees Who Left by Salary Level')
plt.xlabel('Salary Level')
plt.ylabel('Count')
plt.show()
````

````python
plt.figure()
sns.countplot(x = 'Department', hue = 'left', data = df2)
plt.title('Employees Who Left by Salary Level')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
````

````python
# ðŸ“¦ What is a Box Plot?
# âœ… Definition:
# A box plot (or box-and-whisker plot) is a visualization that summarizes the distribution of a numerical variable, grouped by categories.

# ðŸ“Š What it shows:
# Median (middle value)

# Quartiles (Q1 & Q3) â€“ 25% and 75% percentiles

# Interquartile Range (IQR) â€“ height of the box

# Whiskers â€“ extend to data within 1.5 Ã— IQR

# Outliers â€“ points outside the whiskers

# ðŸŽ¯ Purpose:
# See spread, skewness, and outliers in numeric data.

# Compare numerical values across categories (like satisfaction level of employees who left vs stayed).

# âœ… Syntax of Seaborn boxplot
# sns.boxplot(x='category_column', y='numeric_column', data=df)
````

````python
To analyze the relationship between employees who left and their salary levels, the correct visualization approach is **Option 2**:

```python
plt.figure()
sns.countplot(x='salary', hue='left', data=df)
plt.tight_layout()
plt.show()
```

### Explanation:
1. **Why Option 2 is Correct**:
   - **`x='salary'`**: The x-axis represents the salary levels (`low`, `medium`, `high`), allowing you to compare employee counts **across salary brackets**.
   - **`hue='left'`**: The colored bars split each salary group into employees who stayed (`left=0`) vs. those who left (`left=1`). This directly visualizes how attrition rates vary by salary level.
   - You can easily answer questions like:  
     - *"Do employees with lower salaries leave more frequently?"*  
     - *"How does attrition compare between medium and high earners?"*

2. **Why Option 1 is Incorrect for This Analysis**:
   - `x='left'` groups the data by attrition status (left/stayed), with salary as sub-bars. This shows the distribution of salary levels *within* attrition groups but **obscures comparisons between salary levels**.  
   - It answers: *"What salary levels are most common among leavers/stayers?"* but **not** *"How does salary influence attrition?"*

### Key Insights from the Plot (Expected):
- **Lower salary** tiers will likely show a **higher proportion of leavers** (red bars dominate for `salary=low`).
- **Higher salary** tiers will likely show **lower attrition** (blue bars dominate for `salary=high`).

This plot directly highlights how salary impacts employee retentionâ€”critical for HR decision-making.
````
